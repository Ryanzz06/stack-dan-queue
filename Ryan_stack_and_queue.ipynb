```python
#Queue
antrian = []
max_antrian = int(input("masukkan maksimal antrian :"))

def isFull() :
  if len(antrian) == max_antrian :
    print("antrian penuh")
    return True
  return False

def enqueue():
  if isFull() == False :
    r = input("masukkan antrian yang baru : ")
    antrian.append(r)
    print(r, "dimasukkan")
  

def isEmpty() :
  if len(antrian) == 0 :
    print("antrian kosong")
    return True
  return False

def dequeue():
  if isEmpty() == False :
    print(antrian.pop(0), "di hapus")

perintah = 0
while perintah != 5 :
  print('''pilihalah perintah :
  1. enqueue -> menambah antrian
  2. dequeue -> mengahapus antrian
  3. full -> cek antrian apakah penuh/tidam
  4. size -> melihat jumlah perintah
  5. stop''')
  
  perintah = int(input("masukkan perintah :"))
  
  if perintah == 1 :
    enqueue()
  elif perintah == 2 :
    dequeue()
  elif perintah == 3 :
    if isFull() == False :
      print("antrian belum penuh")
  elif perintah == 4 :
    print(" isi antrian saat ini:" , len(antrian),"\n jumlah maksimal antrian:", max_antrian)
```
